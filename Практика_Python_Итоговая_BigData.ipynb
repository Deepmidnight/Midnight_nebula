```python
import pandas as pd


column_names_data = ['user id', 'item id', 'rating', 'timestamp']
column_names_item = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', "'Children's'", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']


ratings = pd.read_csv('u.data.csv', encoding='ISO-8859-1', sep = '\t', names=column_names_data)
movies = pd.read_csv('u.item.csv', encoding='ISO-8859-1', sep = '|', names=column_names_item)
```

# 1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы



```python
ratings.head()
```

```python
movies.head()
```

```python
df_merged = ratings.merge(movies, left_on='item id', right_on='movie id', how='inner')
df_merged.drop('item id', axis=1, inplace=True)
df_merged.sort_values('movie id').head()
```

# 2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок



```python
rating_filter = ratings[['user id', 'rating']]
rating_filter.head()
```

```python
grouped = rating_filter.groupby('user id').count()
grouped.sort_values('rating', ascending=False).head()
```

```python
max_ = grouped['rating'].max()
max_id = grouped[grouped['rating'] == grouped['rating'].max()]
max_id
```

# 3.Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 



```python
user_filter = ratings[['user id', 'rating', 'item id']]
user_filter
```

```python
max_user_filter = user_filter.loc[user_filter['user id'] == 405]
max_user_filter
```

# 4.Добавьте к датафрейму из задания 3 столбцы:
- По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
- Cтолбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей

```python
user_filter_count = max_user_filter
user_filter_count['Count rating'] = ratings.groupby(by='item id')['rating'].transform('count')
user_filter_count['Sum rating'] = ratings.groupby(by='item id')['rating'].transform('sum')
user_filter_count
```

```python
df_user_filter_merged = user_filter_count.merge(movies, left_on='item id', right_on='movie id', how='inner')
df_user_filter_merged.drop('movie id', axis=1, inplace=True)

import numpy as np
df_user_filter_merged['unix date'] = pd.to_datetime(df_user_filter_merged['release date']).astype(np.int64)

df_user_filter_merged = df_user_filter_merged[['user id', 'rating',	'item id',	'Count rating',	'Sum rating',	'release date','unix date', 'Action', 'Adventure', 'Animation', "'Children's'", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
df_user_filter_merged
```

# 5.Сформируйте X_train, X_test, y_train, y_test



```python
X, y = df_user_filter_merged [['Count rating',	'Sum rating',	'unix date', 'Action', 'Adventure', 'Animation', "'Children's'", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']], df_user_filter_merged['rating']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

# 6.Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах





```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

# 7.Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии:



```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train, model.predict(X_train))
```

```python
mean_squared_error(y_test, model.predict(X_test))
```

```python
import findspark
findspark.init()
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()

from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
schema_item = StructType(fields=[
    StructField('movie id', IntegerType()),
    StructField('movie title', StringType()),
    StructField('release date', StringType()),
    StructField('video release date', StringType()),
    StructField('IMDb URL', StringType()),
    StructField('unknown', IntegerType()),
    StructField('Action', IntegerType()),
    StructField('Adventure', IntegerType()),
    StructField('Animation', IntegerType()),
    StructField("'Children's'", IntegerType()),
    StructField('Comedy', IntegerType()),
    StructField('Crime', IntegerType()),
    StructField('Documentary', IntegerType()),
    StructField('Drama', IntegerType()),
    StructField('Fantasy', IntegerType()),
    StructField('Film-Noir', IntegerType()),
    StructField('Horror', IntegerType()),
    StructField('Musical', IntegerType()),
    StructField('Mystery', IntegerType()),
    StructField('Romance', IntegerType()),
    StructField('Sci-Fi', IntegerType()),
    StructField('Thriller', IntegerType()),
    StructField('War', IntegerType()),
    StructField('Western', IntegerType()),
])
df_item=spark.read.csv('u.item.csv', sep = '|', schema = schema_item)
df_item.show(5)
```

```python
schema_data = StructType(fields=[
    StructField("user id", IntegerType()),
    StructField("item id", IntegerType()),
    StructField("rating", IntegerType()),
    StructField("timestamp", IntegerType())
])

df_data=spark.read.csv('u.data.csv', sep='\t', schema = schema_data)
df_data.show(5)
```

# 9.Средствами спарка вывести среднюю оценку для каждого фильма



```python
df_lim_data = df_data.select('item id', 'rating')
df_grp_data=df_lim_data.groupBy('item id')
df_grp_data_avg = df_grp_data.avg('rating')
df_grp_data_avg.sort(['item id'], ascending = True).show(5)
```

# 10.Посчитайте средствами спарка среднюю оценку для каждого жанра



```python
df_total = df_lim_data
df_total = df_total.join(df_item, df_total['item id'] == df_item['movie id'])
df_total.show(2)
df_genre = df_total.select('movie id', 'rating', 'Action', 'Adventure', 'Animation', "'Children's'", 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western')
df_genre.show(5)
#df_genre.write.format('com.databricks.spark.csv').option('header','true').save('df_genre.csv')
```

```python
df_Action = df_genre.select('Action', 'rating').filter("Action=='1'")
df_grp_Action=df_Action.groupBy('Action')
df_grp_Action_avg = df_grp_Action.avg('rating').show()

df_Adventure = df_genre.select('Adventure', 'rating').filter("Adventure=='1'")
df_grp_Adventure=df_Adventure.groupBy('Adventure')
df_grp_Adventure_avg = df_grp_Adventure.avg('rating').show()

df_Animation = df_genre.select('Animation', 'rating').filter("Animation=='1'")
df_grp_Animation=df_Animation.groupBy('Animation')
df_grp_Animation_avg = df_grp_Animation.avg('rating').show()

df_Comedy = df_genre.select('Comedy', 'rating').filter("Comedy=='1'")
df_grp_Comedy=df_Comedy.groupBy('Comedy')
df_grp_Comedy_avg = df_grp_Comedy.avg('rating').show()

df_Crime = df_genre.select('Crime', 'rating').filter("Crime=='1'")
df_grp_Crime=df_Crime.groupBy('Crime')
df_grp_Crime_avg = df_grp_Crime.avg('rating').show()

df_Documentary = df_genre.select('Documentary', 'rating').filter("Documentary=='1'")
df_grp_Documentary=df_Documentary.groupBy('Documentary')
df_grp_Documentary_avg = df_grp_Documentary.avg('rating').show()

df_Drama = df_genre.select('Drama', 'rating').filter("Drama=='1'")
df_grp_Drama=df_Drama.groupBy('Drama')
df_grp_Drama_avg = df_grp_Drama.avg('rating').show()

df_Fantasy = df_genre.select('Fantasy', 'rating').filter("Fantasy=='1'")
df_grp_Fantasy=df_Fantasy.groupBy('Fantasy')
df_grp_Fantasy_avg = df_grp_Fantasy.avg('rating').show()

df_Horror = df_genre.select('Horror', 'rating').filter("Horror=='1'")
df_grp_Horror=df_Horror.groupBy('Horror')
df_grp_Horror_avg = df_grp_Horror.avg('rating').show()

df_Musical = df_genre.select('Musical', 'rating').filter("Musical=='1'")
df_grp_Musical=df_Musical.groupBy('Musical')
df_grp_Musical_avg = df_grp_Musical.avg('rating').show()

```

# 11.В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами (по количеству оценок, либо по самой оценке - на Ваш выбор)



```python
from pyspark.sql.functions import count as _count

df_rate = df_total.select('item id', 'movie title', 'rating')
df_rate_max = df_rate.groupBy('item id','movie title').agg(_count('rating'))
#Датафрейм с 5-ю самыми популярными фильмами по количеству оценок
df_rate_max.sort(['count(rating)'], ascending = False).show(5)
```

```python
#Датафрейм с 5-ю самыми популярными фильмами по суммарной оценке
df_grp_data_sum = df_rate.groupBy('item id','movie title').sum('rating')
df_grp_data_sum.sort(['sum(rating)'], ascending = False).show(5)
```

```python
#Датафрейм с 5-ю самыми непопулярными фильмами по суммарной оценке
df_grp_data_sum = df_rate.groupBy('item id','movie title').sum('rating')
df_grp_data_sum.sort(['sum(rating)'], ascending = True).show(10)
```
